```python
from mp3tagger.training.tools import preprocessing, load_data
import keras
from keras import layers
```

```python
def baseline_model():
    inputs = keras.Input(shape=(14,), name='features')
    x = layers.Dense(10, activation='sigmoid', name='dense_1')(inputs)
    outputs = layers.Dense(3, activation='softmax', name='predictions')(x)
    four_layer = keras.Model(inputs=inputs, outputs=outputs)
    four_layer.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return four_layer
```

```python
df = load_data()
df = preprocessing(df)
```

```python
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import KFold, cross_val_score
from keras import Sequential
from keras.layers import Dense

estimator = KerasClassifier(build_fn=baseline_model, epochs=1, batch_size=1000, verbose=0)
kfold = KFold(n_splits=10, shuffle=True)

results = cross_val_score(estimator, df.values[:, 0:14], df.values[:,13], cv=kfold)
print("Baseline: %.2f%% (%.2f%%)" % (results.mean() * 100, results.std() * 100))
```
